```python
# 計算面積(圓)
# x的最左邊座標和x的最右邊座標
x_left, x_right = -1, 1
# 分成幾分(100)
part = 1000000
# 每一part的x多長
interval = (x_right - x_left) / part

# 總面積
total = 0
# 積分法: 把每一part的面積(小長方形)加總起來
i = 0
while i < part:
    # 每一part的左x座標: -1 + 0.02 * i
    x = x_left + interval * i
    # 算出這裡的y座標多長: 2 * (1 - x ^ 2) ^ 0.5
    y = 2 * (1 - x ** 2) ** 0.5
    # 每一part面積
    area = interval * y
    # 把每一part累積進去
    total = total + area
    i = i + 1
print("積分法估算面積:", total)
```
